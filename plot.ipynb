## Crime Number change with Time from 2006 to 2015

```python
#function to generate data list
def take_x(list):
    x=[]
    for i in range(0,len(list)):
        x.append(int(list[i][0]))
    return x

def take_y(list):
    y = []
    for i in range(0,len(list)):
        y.append(list[i][1])
    return y

def crime_y(crime, population):
    result = []
    for i in range(0,len(crime)):
        result.append(crime[i]/population[i])
    return result
```

```python
#generate data for plot
x_population_year = take_x(population_year)
y_population_year = take_y(population_year)

x_crime_year = take_x(crime_year)
y_crime_year = take_y(crime_year)
y_crime_year = crime_y(y_crime_year, y_population_year)
```

```python
#plot

fig, ax1 = plt.subplots()
fig.set_size_inches(15, 5)
ax1.plot(x_crime_year,y_crime_year, '--o--',color='red',label='Crime Rate')
ax1.set_xlabel('Year')
ax1.set_ylabel('Crime Rate',color='red')

ax2 = ax1.twinx()
ax2.plot(x_population_year, y_population_year,'--o--', color='navy' , label='NYC Population')
ax2.set_ylabel('Population',color = 'navy')

fig.tight_layout()
ax1.legend(loc=2)
ax2.legend(loc=1)
plt.title('NYC Crime Rate and Population change with Year from 2006 to 2016')
plt.show()
```
